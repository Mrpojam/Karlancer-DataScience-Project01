# Task 1

```python
total_transaction_data = [pd.DataFrame(columns=['DAY', 'VALUE']) for i in range (int(max(transaction_data['household_key'].unique()))+1)]
index = 1
tmp_data = pd.DataFrame(columns=['DAY', 'VALUE'])
for i in range(1, 712):
    tmp = pd.DataFrame({'DAY':[i], 'VALUE':[0]})
    tmp_data = tmp_data.append(tmp, ignore_index=True)
tmp_data
```

```python
for i in range(2501):
    total_transaction_data[i] = total_transaction_data[i].append(tmp_data)
```

```python
index = 0
for index, row in transaction_data.iterrows():
    household_key = row['household_key']
    day = row['DAY']
    value = row['SALES_VALUE']
    tmp_data = total_transaction_data[int(household_key)]
    tmp_data[tmp_data['DAY'] == day]['VALUE'] += value
    total_transaction_data[int(household_key)] = tmp_data
    index += 1
    if index % 100000 == 0:
        print (f'index : {index}')
  
```

```python
for i in range(2501):
    if total_transaction_data[i].shape[0] != 0:
        print(total_transaction_data[i].shape)
```
